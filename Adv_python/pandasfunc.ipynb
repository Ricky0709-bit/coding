PANDAS DATA STRUCTURES

```python
import numpy as np 
import pandas as pd
```

1. SERIES

```python
#1. Create a series
data = np.array(['a','b','c'])
s = pd.Series(data)
print(s)
```

```python
#2. Series in index
s = pd.Series(data,index=[100,101,102])
print(s)
```

```python
#3. Key/value objects as series
Calories = {"day 1": 370, "day 2": 390, "day 3": 420}
s = pd.Series(Calories)
print(s)
```

```python
#4. Single item
Calories = {"day 1":  150, "day 2": 170}
x = pd.Series(Calories,index=["day 1"])
print(x)
```
